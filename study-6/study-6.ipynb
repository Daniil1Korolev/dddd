```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("vgsales_10.csv")
df
```

```python
from sklearn.preprocessing import OrdinalEncoder
Oenc = OrdinalEncoder()
Oenc.fit(df[['Platform','Genre','Publisher']])
df
```

```python
Oenc.categories_
```

```python
from sklearn.preprocessing import LabelEncoder
Lenc = LabelEncoder()
Lenc.fit(df['Genre'])
df['Genre'] = Lenc.fit_transform(df['Genre'])
df
```

```python
Lenc.classes_
```

```python
from sklearn.preprocessing import OneHotEncoder
Onhot = OneHotEncoder()
Onhot.fit(df[['Platform']])
df
```

```python
Onhot.categories_
```

```python
Onhot.transform(df[['Platform']]).toarray()
```

```python
from sklearn.preprocessing import Normalizer
Norm = Normalizer()
df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']] = Norm.fit_transform(df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']])
df
```

```python
from sklearn.model_selection import train_test_split
Y = df['Platform']
X = df.drop(columns=['Platform'])
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```
